```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

plt.style.use("fivethirtyeight")

lang_data = pd.read_csv("data.csv")
ids = lang_data['Responder_id']
lang_res = lang_data['LanguagesWorkedWith']

seperator = Counter()

for lang in lang_res:
    seperator.update(lang.split(';'))

language_count = []
language_name = []

for item in seperator.most_common(15):
    language_count.append(item[0])
    language_name.append(item[1])

language_count.reverse()
language_name.reverse()

plt.barh(language_count, language_name)

plt.title("Most Popular Languages") 
plt.xlabel("Number of People Who Use")

plt.tight_layout()
plt.show()
```
